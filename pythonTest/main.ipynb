```python
import matplotlib.pyplot as plt
import numpy as np
import math

# (data.offset * 6 - Math.sin((data.offset - breakPoint[0]) * 5 * Math.PI) * 0.4) * Math.PI

breakPoint = [0.3, 0.4, 0.5]

x=[]
y=[]

for i in np.arange(0, 0.4, 0.005):
    x.append(i)
    if i <= breakPoint[0]:
        y.append(i * 6 * math.pi)
    else:
        print(math.sin((i - breakPoint[0]) * math.pi * 5))
        y.append(i * 6 * math.pi + (math.sin((i - breakPoint[0]) * math.pi * 5 + math.pi / 2) - 1) * math.pi * 0.4)


fig, ax = plt.subplots()
ax.plot(x, y)
ax.plot([0, 0.4], [2 * math.pi, 2 * math.pi])
plt.show()
```
